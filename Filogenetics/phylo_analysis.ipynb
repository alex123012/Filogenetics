```python
def similarity(x, y):
    d = {'G': 0, 'A': 0, 'V': 0, 'L': 0, 'I': 0, 'P': 0, 'M': 0, 'F': 0, 'W': 0,
         'S': 1, 'T': 1, 'C': 1, 'Y': 1, 'N': 1, 'Q': 1,
         'D': 2, 'E': 2,
         'K': 3, 'R':3, 'H': 3, '-': 4}
    res = 0
    for i, j in enumerate(x):
        if j == '-' or y[i] == '-':
            continue
        elif d[j] == d[y[i]]:
            res += 1
    return f'{res}/{len(x)}'
similarity('CCC', 'WQS')
```
